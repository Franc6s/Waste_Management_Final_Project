```python
go_nogo_list = []

for color in waste_data_df["level_"].values:
    if color == "GREEN":
        go_nogo_list.append(0)
    if color == "RED":
        go_nogo_list.append(1)
    if color == "YELLOW":
        go_nogo_list.append(1)

waste_data_df["Pick_Up"] = go_nogo_list

recycler_list = []

for color in waste_data_df["recycler"].values:
    if color == "yes":
        recycler_list.append(1)
    if color == "no":
        recycler_list.append(0)

waste_data_df["recycler"] = recycler_list

waste_data_df = waste_data_df.drop(["level_","description","streamtype"],axis=1)
waste_data_df
```

```python
waste_data_df["date"] = pd.to_datetime(waste_data_df["date"])
waste_data_df = waste_data_df.sort_values(["date"])

test_waste_data_df = pd.DataFrame(waste_data_df)
waste_data_df
```

```python
#last 2500 rows containing data from last 11 days of the month
test_waste_data_df.iloc[-2500:]
```

```python
waste_data_df["date"] = waste_data_df["date"].apply(lambda x : x.toordinal())
waste_data_df
```

```python
time_list = []

for time_string in waste_data_df["time"].values:

    time = time_string.split(" ")[0]
    hours, minutes = time.split(":")
    num_seconds = (int(minutes) * 60)+(int(hours) * 60 * 60)
    time_list.append(num_seconds)

waste_data_df["time"] = time_list

waste_data_df
```

```python
waste_data_df = waste_data_df.sort_values(["date"])
waste_data_df
```

# Machine Learning 

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
import joblib
```

```python
y = waste_data_df["Pick_Up"]
X = waste_data_df.drop(columns=["Pick_Up"])

# Split training/test datasets

X_train = X.iloc[0:-2500].values
X_test = X.iloc[-2500:].values

y_train = y.iloc[0:-2500].values
y_test = y.iloc[-2500:].values
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=78) 

# Fitting the model
rf_model = rf_model.fit(X_train, y_train)

# Evaluate the model
y_pred = rf_model.predict(X_test)
print(f" Random forest model accuracy: {accuracy_score(y_test,y_pred):.3f}")
```
